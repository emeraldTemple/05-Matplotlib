```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data

# Combine the data into a single dataset

# Display the data table for preview
```

```python
cities = pd.read_csv(city_data_to_load)
rides = pd.read_csv(ride_data_to_load)
```

```python
results = pd.merge(rides, cities, on='city', how='inner')
counts = results.groupby(['city']).count()
```

```python
this = rides.groupby(['city']).sum()
cities.sort_values(['city'], axis=0,ascending=True, inplace=True)
cities = cities.reset_index(drop=True)
s = pd.Series(this['fare'].tolist())
cities['total_fare'] = s
counts = rides.groupby(['city']).count()
#all columns are actually the total of rides per city
counts = pd.Series(counts['fare'].tolist())
cities['total_num_rides'] = counts
cities
cities['avg_fare'] = cities['total_fare']/cities['total_num_rides']
index =len(cities)
cities
```

```python
cities['driver_count']
```

```python
driver_pd=cities.groupby(['driver_count'])
drivers = driver_pd.sum()
num_drivers = drivers
num_drivers
```

```python
 def population(i):
        switcher={
                'Urban':'red',
                'Rural':'blue',
                'Suburban':'green'
             }
        return switcher.get(i,"Key-Index Error")
```

## Bubble Plot of Ride Sharing Data

```python
cities.iloc[45].type
```

```python

xaxis = cities['total_num_rides']
yaxis = cities['avg_fare']
z = cities['driver_count']
cdict = {'Urban': 'red', 'Rural': 'blue', 'Suburban': 'green'}

for i in range(index):
    city_type = cities.iloc[i].type
    plt.scatter(xaxis[i], yaxis[i], s=z[i]*9, c = population(city_type), marker='o', edgecolors='black', alpha=0.55)
plt.grid(linestyle='-', linewidth='0.5', color='black')
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
#plt.zlabel('Average Fare ($)')
plt.title('Pyber Ride Sharing Data (2016)')
urban_list = mpatches.Patch(color='red', label='Urban')
suburban_list = mpatches.Patch(color='green', label='Suburban')
rural_list = mpatches.Patch(color='blue', label='Rural')
plt.legend(handles=[urban_list, suburban_list, rural_list])
plt.show()
```

```python
rural_count=cities.groupby(['type'])
rural_count.sum()
all_count = rural_count.sum()
all_count
```

```python
colors = ['#6785d0', '#64a85c', '#b88f3e']
explode = (0, 0.1, 0)
# Create a pie chart
plt.pie(
    # using total Fares
    all_count['total_fare'],
    # City type labels
    labels=['Rural', 'Suburban', 'Urban'],
    # with no shadows
    shadow=True,
    # with colors
    colors=colors,
    # with one slide exploded out
    explode=explode,
    # with the start angle at 90%
    startangle=290,
    # with the percent listed as a fraction
    autopct='%1.1f%%',
    )

# View the plot drop above
plt.axis('equal')
plt.title("% of Total Fares by City Type")

# View the plot
plt.tight_layout()
plt.show()
```

```python
colors = ['#6785d0', '#64a85c', '#b88f3e']
explode = (0, 0.1, 0)
# Create a pie chart
plt.pie(
    # counting drivers in each city
    all_count['total_num_rides'],
    # city type labels
    labels=['Rural', 'Suburban', 'Urban'],
    shadow=True,
    # with colors
    colors=colors,
    # with one slide exploded out
    explode=explode,
    # with the start angle at 90%
    startangle=290,
    # with the percent listed as a fraction
    autopct='%1.1f%%',
    )

# View the plot drop above
plt.axis('equal')
plt.title("% of Total Rides by City Type")

# View the plot
plt.tight_layout()
plt.show()
```
